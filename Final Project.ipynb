```python
df['Education'] =df['Education'].replace({1:"Below College",2:"College",3:"Bachelor",4:"Master",5:"Doctrate"})
df['EnvironmentSatisfaction'] = df['EnvironmentSatisfaction'].replace({1:"Low",2:"Medium",3:"High",4:"Very High"})
df['JobInvolvement'] = df['JobInvolvement'].replace({1:"Low",2:"Medium",3:"High",4:"Very High"})
df['JobSatisfaction'] = df['JobSatisfaction'].replace({1:"Low",2:"Medium",3:"High",4:"Very High"})
df['JobLevel'] = df['JobLevel'].replace({1:"Entry Level",2:"Junior Level",3:"Mid Level",4:"Senior Level",5:"Executive Level"})
df['PerformanceRating'] = df['PerformanceRating'].replace({1:"Low",2:"Good",3:"Excellent",4:"Outstanding"})
df['RelationshipSatisfaction'] = df['RelationshipSatisfaction'].replace({1:"Low",2:"Medium",3:"High",4:"Very High"})
df['WorkLifeBalance'] = df['WorkLifeBalance'].replace({1:"Bad",2:"Good",3:"Better",4:"Best"})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Checking the correlation through heatmap
plt.figure(figsize = (12,8))
sns.heatmap(df_num.corr(numeric_only=True), annot=True)
```

```python
# Set up the figure and axes for subplots
fig, axes = plt.subplots(4, 4, figsize=(20, 20))

# Flatten the axes array for easier iteration
axes = axes.flatten()

features = ['Attrition', 'BusinessTravel', 'Department', 'Education',
            'EnvironmentSatisfaction', 'Gender', 'JobInvolvement', 'JobLevel', 'JobSatisfaction', 'MaritalStatus', 'OverTime',
            'RelationshipSatisfaction', 'StockOptionLevel', 'WorkLifeBalance','YearsInCurrentRole', 'YearsSinceLastPromotion']

# Plot each feature
for i, feature in enumerate(features):
    ax = sns.countplot(data=df, x=feature,hue='Attrition', ax=axes[i], palette ='husl')
    axes[i].set_title(f'{feature} Count')
plt.show()
```

```python
# Creating a column for Age Group

bins = [18, 25, 32, 39, 46, 53, 60]
labels = ['18-24', '25-31', '32-38', '39-45', '46-52', '53-60']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
# Visualizing Overall Attrition

sns.countplot(x=df['Attrition'])
sns.set_palette("husl")
plt.xlabel("Attrition")
plt.ylabel("Count")
plt.title("Distribution of Attrition")
for p in plt.gca().patches:
    height = p.get_height()
    plt.text(p.get_x() + p.get_width() / 2., height, f'{height}',ha='center', va='bottom')
plt.show()
```
